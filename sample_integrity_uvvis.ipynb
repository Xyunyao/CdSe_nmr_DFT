## UV-VIS spectra to show sample integrity after DNP experiments

```python
data_path="/Users/yunyao_1/Dropbox/CdSe-data-owen/Current/revision/CdSe350_BeforeAfterNMR.csv"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv(data_path,header=1)
data = data.iloc[:, 3:]  # Keep columns starting from index 3
data.columns = ['Wavelength', 'PreNMR', 'PostNMR']  # Rename the remaining columns

data.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Wavelength'], data['PreNMR'], label='Pre-NMR')
plt.plot(data['Wavelength'], data['PostNMR'], label='Post-NMR')
plt.axvline(x=350, color='red', linestyle='--', linewidth=1, label='350 nm')
plt.xlim(285, 650)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.legend()
plt.show()  
```
